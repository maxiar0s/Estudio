# **Función range**

La función range devuelve una sucesión de números enteros

*   **range(n)**:devolverá una sucesión de números 
enteros de 0 a n-1

    Ejemplo: range(5) devolverá 0, 1, 2, 3, 4
*   **range(a,b)**:devolverá una sucesión de números 
enteros de a hasta b-1

    Ejemplo:  range(3,9) devolverá 3,4,5,6,7,8

*   **range(a,b,c)**:devolverá una sucesión de números 
enteros de a hasta b-1 con incremento de c

    Ejemplo:  range(3,12,4) devolverá 3,7,11
