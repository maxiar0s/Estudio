````python
#Ejemplo1: Ingresar un número entero (validar) y luego mostrarlo
while(True):
	try:
		num=int(input("Ingrese un número entero: "))
		break
	except ValueError:
		print("\n\tDebe ser un número entero!!\n")

print("\nEl número ingresado es "+str(num))
````

````python
#Ejemplo 2: Ingresar un número entero en el rango 18 a 65 (validar) y luego mostrar si es par o impar
while(True):
	try:
		num=int(input("Ingrese un número entero en el rango de 18 a 65: "))
		if(num>=18 and num<=65):
			break
		else:
			print("\n\tEl número debe estar en el rango 18 a 65\n")
	except ValueError:
		print("\n\tDebe ser un número entero!!\n")

if(num%2==0):
	print("\nEl número es par")
else:
	print("\nEl número es impar")
````
