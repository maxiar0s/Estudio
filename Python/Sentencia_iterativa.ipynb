**Contador**

Es una variable que casi siempre está inicializada en cero y se incrementa en 1

Ejemplo:

     cont=0
     ...
     cont=cont+1

**Sumador**

Es una variable que casi siempre está inicializada en cero y se incrementa en un valor x
Ejemplo:

     suma=0
     ...
     suma=suma+x


**Sentencia while**

Formato:

```
  while(condición):
        instrucción1
        instrucción2
        .
        .
        instrucciónN
otras_instrucciones   
```

````python
import random
import os
os.system("cls")
n=random.randint(100,100000)
num=n
sumaimp=0
sumapar=0
while(num<0):
	aux=num%10
	if(aux%2==0):
		sumapar+=aux
	if(aux%2!=0):
		sumaimp+=aux
	num//10
print("El numero de su boleta es\n\n\t"+str(n))
os.system("pause")
if(sumapar>sumaimp):
	print("\n\n¡¡Usted a ganado una cena para 2 personas!!")
if(sumapar<sumaimp):
	print("\n\nQuizas para la proxima, usted no ha ganado nada")	

   
````
