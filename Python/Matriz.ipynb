# **Creación de una matriz**

**Se pueden crear vacias:**

1.   import numpy

     nombre_matriz=numpy.empty([cantFilas,cantColumnas])

2.   import numpy

     nombre_matriz=numpy.empty([cantFilas,cantColumnas],type(tipoalmacenar))

```python
#Ejemplo 1
import numpy
mat=numpy.empty([2,3])
print(mat)
for i in range(2):
  print("Alumno "+str(i+1))
  for j in range(3):
    mat[i][j]=float(input("nota "+str(j+1)+": "))
print("Los notas ingresados son: ")
print(mat)
```

```python
#Ejemplo 2
import numpy
mat=numpy.empty([2,3],type(str))
print(mat)
for i in range(2):
  for j in range(3):
    mat[i][j]=input("Nombre: ")
print("Los nombres ingresados son: ")
print(mat)
```

# **Matrices cuadradas**

Son aquellas que tienen la misma cantidad de filas y columnas. Una matriz Cuadrada posee diagonal principal y secundaria

matCua.png

```python
"""
Llenar una matriz de orden NxN con dígitos (aletorio) y luego:
- Mostrar toda la matriz
- Mostrar la suma de los elemtos de la diagonal principal
- Mostrar cantidad de elementos menores a 5 de la diagonal secundaria 
"""
import numpy
import random
n=int(input("Ingrese el orden de la matriz: "))
digitos=numpy.empty([n,n],type(int))

for i in range(n):
  for j in range(n):
    digitos[i][j]=random.randint(0,9)

print("\nLa matriz es:")
print(digitos)

suma=0
for i in range(n):
  for j in range(n):
    if(i==j):
      suma+=digitos[i][j]
print("\nSuma de los elementos de la DP: "+str(suma))

cont=0
for i in range(n):
  for j in range(n):
    if(i+j==n-1):
      if(digitos[i][j]<5):
        cont+=1
print("\nCantidad de los elementos menor a 5 de la DS: "+str(cont))
```
