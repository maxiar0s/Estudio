**Creación de un arreglo**
Se pueden crear de dos formas:

1. import numpy

   nombre_arreglo=numpy.array([])


2. from numpy import array

   nombre_arreglo=[]
   

```python
"""
Ejemplo 1:
Realizar un programa que ingrese nombres hasta que el usuario desee 
y luego mostrar todos los nombres ingresados
"""
from numpy import array
nombres=[]

resp="S"
while(resp=="S"):
  nom=input("Nombre: ")
  nombres.append(nom)
  resp=input("\n¿Desea ingresar otro nombre S/N?: ").upper()


print("\n\tLos nombres ingresados son")
i=0
while(i<len(nombres)):
  print(nombres[i],end=" ")
  i+=1

```

```python
"""
Ejemplo 2:
Realizar un programa que permita almacenar 20 números enteros en forma 
aleatoria (en el rango 0 a 15) en un arreglo y luego:
a. Dejar en otro arreglo sólo los pares 
b. Mostrar ambos arreglos
c. Mostar la cantidad de elementos del primer arreglo que coinciden con el índice
"""
from numpy import array
import random
numeros=[]
pares=[]

i=0
while(i<20):
  numeros.append(random.randint(0,15))
  i+=1

p=0
while(p<20):
  if(numeros[p]%2==0):
    pares.append(numeros[p])
  p+=1

print("\n\tLos numeros generados son")
i=0
while(i<20):
  print(numeros[i],end=" ")
  i+=1

print("\n\tSólo los pares")
i=0
while(i<len(pares)):
  print(pares[i],end=" ")
  i+=1

i=0
cont=0
while(i<20):
  if(numeros[i]==i):
    cont+=1
  i+=1
print("\nLa cantidad de elementos que coiciden con el índice son "+str(cont))

```
