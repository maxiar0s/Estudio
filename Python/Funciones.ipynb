# **Funciones**

   Una función es un conjunto de instrucciones reutilizable que se puede ejecutar muchas veces dentro del programa

Formato:
```
def nombre_funcion(parámetros de entrada):
    instrucciones
    return (valor)

Ejemplo:
Realizar un programa que permita:
- Llenar dos arreglos de tamaño n y m con 
números enteros (en forma aleatoría en el rango 10 a 50)
- Mostrar los arreglos 
- Mostrar cantidad de elementos pares de ambos arreglos
- Dejar en otro arreglo (impares1) los elementos impares del primero 
- Dejar en otro arreglo (impares2) los elementos impares del segundo 
- Mostrar los arrglos con los impares, si no tienen elementos, mostrar mensaje adecuado


````python
#Sin funciones
from numpy import array
import random

n=int(input("Ingrese cantidad de elementos del primer arreglo: "))
m=int(input("Ingrese cantidad de elementos del primer arreglo: "))

arre1=[]
arre2=[]

for i in range(n):
	arre1.append(random.randint(10,50))

for i in range(m):
	arre2.append(random.randint(10,50))

print("\n\tElementos del arreglo1: ")
print(arre1)

print("\n\tElementos del arreglo2: ")
print(arre2)

cont=0
for i in range(n):
	if(arre1[i]%2==0):
		cont+=1
print("\nCantidad de elementos pares del primero: "+str(cont))

cont=0
for i in range(m):
	if(arre2[i]%2==0):
		cont+=1
print("\nCantidad de elementos pares del segundo: "+str(cont))

imp1=[]
for i in range(n):
  if(arre1[i]%2!=0):
    imp1.append(arre1[i])

if(len(imp1)!=0):
	print("\nLos elementos impares del primero son:")
	print(imp1)
else:
	print("\nEl primero no tiene elementos impares")

imp2=[]
for i in range(m):
  if(arre2[i]%2!=0):
   imp2.append(arre2[i])

if(len(imp2)!=0):
	print("\nLos elementos impares del segundo son:")
	print(imp2)
else:
	print("\nEl segundo no tiene elementos impares")
````
