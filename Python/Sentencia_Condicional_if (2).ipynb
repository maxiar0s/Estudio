# **Sentencia Condicional if**

**Formato:** Esta sentencia se puede utilizar de tres formas:
1. Sentencia if  
```
  if(condición):
        instrucción1
        instrucción2
        .
        .
        instrucciónIfN
otras_instrucciones   
```

````python
#Ejemplo1: programa que muestra el total a pagar de un producto, si el precio es menor o igual a 10000
#se le realizará un descunto de un 25% del precio
desc=0
valor=int(input("Ingrese precio del producto "))
if(valor<=10000):
  desc=valor*0.25
  print("El valor es menor o igual a $10000")
  print("Se le realizará un descuento de un 25%")
total=valor-desc
print("El total a pagar es "+str(total))
````


2.  Sentencia if-else
``` 
  if(condición):
        instrucciónIf1
        instrucciónIf2
        .
        .
        instrucciónIfN
  else:
        instrucciónElse1
        instrucciónElse2
        .
        .
        instrucciónElseN 
otras_instrucciones   
```

````python
#Ejemplo2:programa que muestra el total a pagar de un producto, si el precio es menor o igual a 10000
#se le realizará un descunto de un 25% del precio
valor=int(input("Ingrese precio del producto "))
if(valor<=10000):
  desc=valor*0.25
  print("El valor es menor o igual a $10000")
  print("Se le realizará un descuento de un 25%")
else:
  desc=0
total=valor-desc
print("El total a pagar es "+str(total))
````

3.  Sentencia if-elif-else
``` 
  if condición_1:
        bloque 1
  elif condición_2:
        bloque 2
  else:
        bloque 3
```
Es equivalente a:
``` 
  if condición_1:
        bloque 1
  else
        if condición_2:
            bloque 2
        else:
            bloque 3
```
En diagrama tenemos:
elsi.png

````python
#Ejemplo3: programa que muestra el total a pagar de un producto, si el precio es menor o igual a 10000
#se le realizará un descunto de un 25% del precio; si es mayor a 10000 y menor que 50000, será de un 35%
valor=int(input("Ingrese precio del producto "))
if(valor<=10000):
  desc=valor*0.25
  print("El valor es menor o igual a $10000")
  print("Se le realizará un descuento de un 25%")
elif(valor>10000 and valor<50000):
  desc=valor*0.35
  print("El valor es mayor a $10000 y menor a $50000")
  print("Se le realizará un descuento de un 35%")
else:
  desc=0
total=valor-desc
print("El total a pagar es "+str(total))
````
